```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.read_csv("supermarket_sales2.csv")
```

```python
df.dropna(inplace=True)
# to drop/delete the rows in the df with null values
df.head()#to print the first rows to see check the deleted rows
```

```python
df.fillna(df['Unit price'].mean(), inplace=True)
df.head()
#upload the file again and run the above comands , this it will find the mean of the non null values and replace the null values with the mean value
```

```python
sns.histplot(df['Unit price'])
plt.show()
#gives the histogram (graphical reprssentation) of the unit price column
```

```python
sns.boxplot(x='Payment', y='Unit price', data=df)
plt.show()
#gives the boxplot (graphical reprssentation) of the unit price column
# Ewet is the typing error, u need to save th efile again and run the entire thing again
```

```python
sns.scatterplot(x='cogs', y='Unit price', data=df)
plt.show()
#gives the scatterplot (graphical reprssentation) of the unit price column
```

```python
sns.countplot(x='Branch' , data=df)
plt.show()
#gives the countplot (graphical reprssentation) of the unit price column
#bar charts = visualise the distribution of categorial variables
```

```python
af=df[['Unit price', 'Quantity', 'Tax 5%', 'cogs', 'gross income', 'Rating']]
corr=af.corr()
sns.heatmap(corr, annot=True)
plt.show()
#gives the heatmap (graphical reprssentation) of the unit price column
#correlation matrix = understand the relations between multiple numerical values
```
